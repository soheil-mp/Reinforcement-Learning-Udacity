```python
# Import the libraries
import numpy as np
import gym
import sys
from collections import defaultdict
from plot_utils import plot_blackjack_values, plot_policy
```

```python
# Create an instance of environment
env = gym.make("Blackjack-v0")
```

```python
# Print the state space
print("State Space: \n", env.observation_space)
```

```python
# Print the action space
print("Action Space: \n", env.action_space)
```

```python
### Play BlackJack with random policy

# Iterate through number of episodes
for i_episode in range(3):
    
    # Reset the state
    state = env.reset()
    
    print("=================================")
    # Infinite loop
    while True:
        
        # Print the current state
        print("Current state: ", state)
        
        # Sample a action uniformely random
        action = env.action_space.sample()
        
        # Run the one step dymanics of environment
        state, reward, done, info = env.step(action)
        
        # If terminal state
        if done:
            
            # Print the infos
            print("Game Ended!")
            print("Reward: ", reward)
            print("YOU WON!") if reward > 0 else print("YOU LOST!")
            print("=================================")
            
            # Break 
            break
```

```python
# Function for generating episode
def generate_episode(blackjack_env):
    """
    Generate episode.
    
    """
    # Initialize an empty array for episodes
    episodes = []
    
    # Reset the state
    state = env.reset()
    
    # Infinite loop
    while True:
        
        # Choose action
        #        - Choose STICK if sum is over 18
        #        - Choose HIT if sum is 18 or less
        action = 0 if state[0] > 18 else 1
        
        # Run one-step dynamics of environment
        next_state, reward, done, info = env.step(action)
        
        # Append S, A, R to episodes
        episodes.append((state, action, reward))
        
        # Update the state
        state = next_state
        
        # Break if terminal state
        if done:
            break
            
    return episodes
```

```python
# Play BlackJack for 3 times
for i_episode in range(3):
    print(generate_episode(env))
```

```python
# Monte Carlo Prediction (first visit)
def mc_prediction(blackjack_env, generate_episode, num_episodes, gamma = 1.0):
    
    # Initialize returns
    returns = defaultdict(list)
    
    # Loop over episodes
    for i_episode in range(1, num_episodes+1):
        
        # Print the episodes
        if i_episode % 1000 == 0:
            print("\rEpisode: {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Generate an episode
        episode = generate_episode(blackjack_env)
        
        # Get the S, A, R
        states, actions, rewards = zip(*episode)
        
        # Prepare for discouting
        discounts = np.array([gamma**i for i in range(len(rewards)+1)])
        
        # Iterate through states
        for i, i_state in enumerate(states):
            
            # Store the return for each visit
            returns[i_state].append(sum(rewards[i:]*discounts[:-(1+i)]))
            
    # Calcualte the state-value function
    V = {k: np.mean(v) for k, v in returns.items()}
    
    return V
```

```python
# Get the state-value function
V = mc_prediction(env, generate_episode, 500000)
```

```python
# Visualize the state-value function
plot_blackjack_values(V)
```

```python
# Generate stochastic episodes
def generate_episode_from_limit_stochastic(blackjack_environment):
    """
    Generate stochastic episodes
    """
    # Initialize an empty array for episodes
    episode = []
    
    # Reset the state
    state = blackjack_environment.reset()
    
    # Infinite loop
    while True:
        
        # Define the probabilities for action
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        
        # Choose the action based on the probabilities
        action = np.random.choice(np.arange(2), p = probs)
        
        # Run the one-step dynamicsc of environment
        next_state, reward, done, info = blackjack_environment.step(action)
        
        # Append the S, A, R
        episode.append((state, action, reward))
        
        # Update the state
        state = next_state
        
        # Break if in terminal state
        if done:
            break
            
    return episode
```

```python
# First-visit MC prediction for action values
def mc_prediction_q(blackjack_environment, episode_generator, num_episodes, gamma = 1.0):
    """
    First-visit MC prediction for action values
    """
    # Initialize the return
    returns_sum = defaultdict(lambda: np.zeros(blackjack_environment.action_space.n))
    
    # Initialize the number of visits
    N = defaultdict(lambda: np.zeros(blackjack_environment.action_space.n))
    
    # Initialize the action-value function
    Q = defaultdict(lambda: np.zeros(blackjack_environment.action_space.n))
    
    # Iterate through number of episodes
    for i_episode in range(1, num_episodes+1):
        
        # Print the episode numbers
        if i_episode % 1000 == 0:
            print("\rEpisode: {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Generate episode
        episode = episode_generator(blackjack_environment)
        
        # Get the S, A, R
        states, actions, rewards = zip(*episode)
        
        # Get the discount
        discounts = np.array([gamma**i for i in range(len(rewards)+1)])
        
        # Iterate states
        for index, i_state in enumerate(states):
            
            # Update the retunrs sum
            returns_sum[i_state][actions[index]] += sum(rewards[index:]*discounts[:-(index+1)])
            
            # Update the number of visits
            N[i_state][actions[index]] += 1.0
            
            # Update the action-value function
            Q[i_state][actions[index]] = returns_sum[i_state][actions[index]] / N[i_state][actions[index]]
        
    return Q
```

```python
# Get the action-value function
Q = mc_prediction_q(env, generate_episode_from_limit_stochastic, 500000)

# Get the state-value funtion
V_to_plot = dict((k, (k[0]>18)*(np.dot([0.8, 0.2], v)) + (k[0]<=18)*(np.dot([0.2, 0.8], v))) for k, v in Q.items())

# Plot the state-value function
plot_blackjack_values(V_to_plot)
```

## GLIE

```python
# Function for generating episode
def generate_episode_from_Q(env, Q, epsilon, nA):
    """
    Generate an episode from following the epsilon-greedy policy.
    """
    # Initialize an empty array for episodes
    episode = []
    
    # Reset the state
    state = env.reset()
    
    # Infinite loop
    while True:
        
        # Choose an action
        action = np.random.choice(a = np.arange(nA), p = get_probs(Q_s = Q[state], epsilon = epsilon, nA = nA)) \
                 if state in Q else env.action_space.sample()
        
        # Run one-step dynamics of environment
        next_state, reward, done, info = env.step(action)
        
        # Append S, A, R
        episode.append((state, action, reward))
        
        # Update the state
        state = next_state
        
        # Break if terminal state
        if done:
            break
            
    return episode
```

```python
# Function for getting the probabilities for epsilon-greedy policy
def get_probs(Q_s, epsilon, nA):
    """
    Get the probabilities for epsilon-greedy policy
    """
    # Get the policy
    policy_s = np.ones(nA) * epsilon / nA
    
    # Choose the best action
    best_a = np.argmax(Q_s)
    
    # Update the policy
    policy_s[best_a] = 1 - epsilon + (epsilon / nA)
    
    return policy_s
```

```python
# Function for updating the action-value function using the most recent episode
def update_Q_GLIE(env, episode, Q, N, gamma):
    """
    Update the action-value function using the most recent episode
    """
    # Retrieve the S, A, R
    states, actions, rewards = zip(*episode)
    
    # Prepare the dicounting
    discounts = np.array([gamma**i for i in range(len(rewards)+1)])
    
    # Iterate through states
    for index, i_state in enumerate(states):
        
        # Get the old action value
        Q_old = Q[i_state][actions[index]]
        
        # Get the old number of visits
        N_old = N[i_state][actions[index]]
        
        # Update the action value
        Q[i_state][actions[index]] = Q_old + (sum(rewards[index:]*discounts[:-(index+1)]) - Q_old) / (N_old + 1)
        
        # Update the number of visits
        N[i_state][actions[index]] += 1
        
    return Q, N
```

```python
# Function for Monte-Carlo (MC) Control with greedy in the limit with infinite exploration (GLIE)
def mc_control_GLIE(env, num_episodes, gamma = 1.0):
    """
    Monte-Carlo (MC) Control with greedy in the limit with infinite exploration (GLIE)
    """
    # Number of actions in action space
    nA = env.action_space.n
    
    # Initialize the action-value function with zeros
    Q = defaultdict(lambda: np.zeros(nA))
    
    # Initializ ethe number of visits
    N = defaultdict(lambda: np.zeros(nA))
    
    # Iterate through episodes
    for i_episode in range(1, num_episodes+1):
        
        # Print the episode number
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
            
        # Set the epsilon
        epsilon = 1.0 / ((i_episode/8000)+1)
        
        # Generate episode using epsilon-greedy policy
        episode = generate_episode_from_Q(env, Q, epsilon, nA)
        
        # Update the action-value function
        Q, N = update_Q_GLIE(env, episode, Q, N, gamma)
        
    # Determine the policy corresponding to the last action-value function
    policy = dict((k, np.argmax(v)) for k, v in Q.items())
    
    return policy, Q
```

```python
# Get the optimal policy and action-value function
policy_glie, Q_glie = mc_control_GLIE(env, 500000)
```

```python
# Get the state-value function
V_glie = dict((k, np.max(v)) for k, v in Q_glie.items())
```

```python
# Visualize the state-value function
plot_blackjack_values(V_glie)
```

```python
# Visualize the policy
plot_policy(policy_glie)
```

```python
# Update the action-value function using the most recent episode
def update_Q_alpha(env, episode, Q, alpha,gamma):
    """
    Update the action-value function using the most recent episode
    """
    # Get the S, A, R
    states, actions, rewards = zip(*episode)
    
    # Prepare for discounting
    discounts = np.array([gamma**i for i in range(len(rewards)+1)])
    
    # Iterate through states
    for index, i_state in enumerate(states):
        
        # Get the old action-value function 
        Q_old = Q[i_state][actions[index]]
        
        # Update the action-value function
        Q[i_state][actions[index]] = Q_old + alpha * (sum(rewards[index:] * discounts[:-(index+1)]) - Q_old)
        
    return Q
```

```python
# Function for MC Control Constant Alpha 
def mc_control_alpha(env, num_episodes, alpha, gamma = 1.0):
    """
    MC Control Constant Alpha 
    """
    # Get the number of actions in actions space
    nA = env.action_space.n
    
    # Initialize the action-value function
    Q = defaultdict(lambda: np.zeros(nA))
    
    # Iterate through episodes
    for i_episode in range(1, num_episodes+1):
        
        # Print the episode number
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
            
        # Set the epsilon
        epsilon = 1.0 / ((i_episode/8000)+1)
        
        # Generate an episode using epsilon-greedy policy
        episode = generate_episode_from_Q(env, Q, epsilon, nA)
        
        # Update the action-value function
        Q = update_Q_alpha(env, episode, Q, alpha, gamma)
        
    # Determine the policy corresponding to the final action-value function estimate
    policy = dict((k, np.argmax(v)) for k, v in Q.items())
    
    return policy, Q
```

```python
# Get the optimal policy and action-value function
policy_alpha, Q_alpha = mc_control_alpha(env, 500000, 0.02)
```

```python
# Get the state-value function
V_alpha = dict((k, np.max(v)) for k, v in Q_alpha.items())
```

```python
# Visualize the state-value function
plot_blackjack_values(V_alpha)
```

```python
# Visualize the policy
plot_policy(policy_alpha)
```
