# 1. Explore the MountainCar Environment 

<hr>

In this jupyter notebook, we will use a environment in [OpenAI Gym](https://gym.openai.com/) called MountainCar. This environment has:
- Continous state space 
- Discrete action space.

````python
# Initialize MountainCar environment
env = gym.make('MountainCar-v0')
````

````python
# Set a random seed
env.seed(505);
````

````python
### Play with random actions

# Resets the state and get an initial observation
state = env.reset()

# Initialize score to 0
score = 0

# Iterate through the time-steps
for t in range(200):
    
    # Get a random action 
    action = env.action_space.sample()
    
    # Render the environment
    env.render()
    
    # Run one timestep of the environment's dynamics for getting the state and the reward
    state, reward, done, _ = env.step(action)
    
    # Apend the reward to score
    score += reward
    
    # If terminal state then break
    if done:
        break 
        
# Print the final score
print('Final score:', score)

# Close the environment
env.close()
````

# 2. Discretize the State Space with a Uniform Grid

<hr>

We will discretize the space using a uniformly-spaced grid. Implement the following function to create such a grid, given the lower bounds (`low`), upper bounds (`high`), and number of desired `bins` along each dimension. It should return the split points for each dimension, which will be 1 less than the number of bins.

For instance, if `low = [-1.0, -5.0]`, `high = [1.0, 5.0]`, and `bins = (10, 10)`, then your function should return the following list of 2 NumPy arrays:

```
[array([-0.8, -0.6, -0.4, -0.2,  0.0,  0.2,  0.4,  0.6,  0.8]),
 array([-4.0, -3.0, -2.0, -1.0,  0.0,  1.0,  2.0,  3.0,  4.0])]
```

Note that the ends of `low` and `high` are **not** included in these split points. It is assumed that any value below the lowest split point maps to index `0` and any value above the highest split point maps to index `n-1`, where `n` is the number of bins along that dimension.

````python
def create_uniform_grid(low, high, bins=(10, 10)):
    """
    Define a uniformly-spaced grid that can be used to discretize a space.
    
    PARAMETERS
    ------------------------
        - low : Lower bounds for each dimension of the continuous space. (array_like)
        - high : Upper bounds for each dimension of the continuous space. (array_like)
        - bins : Number of bins along each corresponding dimension. (tuple)
    
    RETURNS
    ------------------------
        - grid : A list of arrays containing split points for each dimension. (list of array_like)
    """
    # Create a uniformly-spaced grid
    grid = [np.linspace(low[dim], high[dim], bins[dim] + 1)[1:-1] for dim in range(len(bins))]
        
    return grid
````

````python
# Define low, high, and bins
low = [-1.0, -5.0]
high = [1.0, 5.0]
bins = (10, 10)

# Get the uniformly-spaced grid
grid = create_uniform_grid(low, high)

# Print the grid
print("Grid: \n", grid)

# Print the process
print("\nProcess: [<low>, <high>] / <bins> => <splits>")
for l, h, b, splits in zip(low, high, bins, grid):
    print("[{}, {}] / {} => {}".format(l, h, b, splits))
````

Now write a function that can convert samples from a continuous space into its equivalent discretized representation, given a grid like the one you created above. You can use the [`numpy.digitize()`](https://docs.scipy.org/doc/numpy-1.9.3/reference/generated/numpy.digitize.html) function for this purpose.

Assume the grid is a list of NumPy arrays containing the following split points:
```
[array([-0.8, -0.6, -0.4, -0.2,  0.0,  0.2,  0.4,  0.6,  0.8]),
 array([-4.0, -3.0, -2.0, -1.0,  0.0,  1.0,  2.0,  3.0,  4.0])]
```

Here are some potential samples and their corresponding discretized representations:
```
[-1.0 , -5.0] => [0, 0]
[-0.81, -4.1] => [0, 0]
[-0.8 , -4.0] => [1, 1]
[-0.5 ,  0.0] => [2, 5]
[ 0.2 , -1.9] => [6, 3]
[ 0.8 ,  4.0] => [9, 9]
[ 0.81,  4.1] => [9, 9]
[ 1.0 ,  5.0] => [9, 9]
```

**Note**: There may be one-off differences in binning due to floating-point inaccuracies when samples are close to grid boundaries, but that is alright.

````python
def discretize(sample, grid):
    """
    Discretize a sample as per given grid.
    
    PARAMETERS
    ---------------------
        - sample : A single sample from the (original) continuous space. (array_like)
        - grid : A list of arrays containing split points for each dimension. (list of array_like)
    
    RETURNS
    ---------------------
        - discretized_sample : A sequence of integers with the same number of dimensions as sample. (array_like)
    """
    # Discretize the given sample; apply this along each dimension
    discretized = list(int(np.digitize(s, g)) for s, g in zip(sample, grid)) 
    
    return discretized
````

````python
# Define low, high, and bins
low = [-1.0, -5.0]
high = [1.0, 5.0]
bins = (10, 10)

# Get the uniformly-spaced grid
grid = create_uniform_grid(low, high)

# Get samples
samples = np.array(
    [[-1.0 , -5.0],
     [-0.81, -4.1],
     [-0.8 , -4.0],
     [-0.5 ,  0.0],
     [ 0.2 , -1.9],
     [ 0.8 ,  4.0],
     [ 0.81,  4.1],
     [ 1.0 ,  5.0]])

# Apply discretization along each dimension
discretized_samples = np.array([discretize(sample, grid) for sample in samples])

# Print the samples and its discretized form
print("Samples:", repr(samples), sep="\n")
print("\nDiscretized samples:", repr(discretized_samples), sep="\n")
````

# 3. Visualization Original and Discretized Samples

<hr>

It might be helpful to visualize the original and discretized samples to get a sense of how much error you are introducing.

````python
def visualize_samples(samples, discretized_samples, grid, low = None, high = None):
    """
    Visualize original and discretized samples on a given 2-dimensional grid.
    
    ARGUMENTS
    ----------------------
        - samples: Sample from the (original) continuous space. 
        - discretized_samples: Discretized representation of given samples
        - grid: A list of arrays containing split points for each dimension. 
        - low : Lower bounds for each dimension of the continuous space. (array_like)
        - high : Upper bounds for each dimension of the continuous space. (array_like)
    """
    # Create a figure and a set of subplots
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Show grid
    ax.xaxis.set_major_locator(plt.FixedLocator(grid[0]))
    ax.yaxis.set_major_locator(plt.FixedLocator(grid[1]))
    ax.grid(True)
    
    # If bounds (low, high) are specified, use them to set axis limits
    # Otherwise use first, last grid locations as low, high (for further mapping discretized samples)
    if low is not None and high is not None:
        ax.set_xlim(low[0], high[0])
        ax.set_ylim(low[1], high[1])
    else:
        low = [splits[0] for splits in grid]
        high = [splits[-1] for splits in grid]

    # Map each discretized sample (which is really an index) to the center of corresponding grid cell
    grid_extended = np.hstack((np.array([low]).T, grid, np.array([high]).T))  # add low and high ends
    grid_centers = (grid_extended[:, 1:] + grid_extended[:, :-1]) / 2  # compute center of each grid cell
    locs = np.stack(grid_centers[i, discretized_samples[:, i]] for i in range(len(grid))).T  # map discretized samples

    # Plot
    ax.plot(samples[:, 0], samples[:, 1], 'o')  # plot original samples
    ax.plot(locs[:, 0], locs[:, 1], 's')  # plot discretized samples in mapped locations
    ax.add_collection(mc.LineCollection(list(zip(samples, locs)), colors='orange'))  # add a line connecting each original-discretized sample
    ax.legend(['original', 'discretized'])
````

````python
# Visualize the discretized samples
visualize_samples(samples, discretized_samples, grid, low, high)
````

Now that we have a way to discretize a state space, let's apply it to our reinforcement learning environment.

````python
### Discretize the MountainCar evnironment

# Create a grid
state_grid = create_uniform_grid(env.observation_space.low, env.observation_space.high, bins=(10, 10))

# Get samples from the space
state_samples = np.array([env.observation_space.sample() for i in range(10)])

# Apply discretization along each dimension
discretized_state_samples = np.array([discretize(sample, state_grid) for sample in state_samples])

# Visualize
visualize_samples(state_samples, discretized_state_samples, state_grid,
                  env.observation_space.low, env.observation_space.high)
plt.xlabel('position')
plt.ylabel('velocity')
plt.show()
````

You might notice that if you have enough bins, the discretization doesn't introduce too much error into your representation.  So we may be able to now apply a reinforcement learning algorithm (like Q-Learning) that operates on discrete spaces.  Give it a shot to see how well it works!

# 4. Q-Learning Agent

<hr>

Provided below is a simple Q-Learning agent. Implement the `preprocess_state()` method to convert each continuous state sample to its corresponding discretized representation.

````python
# Q-Learning agent class
class QLearningAgent:
    """
    Q-Learning agent that can act on a continuous state space by discretizing it.
    
    """
    
    # Init function
    def __init__(self, env, state_grid, alpha = 0.02, gamma = 0.99, epsilon = 1.0, epsilon_decay_rate = 0.9995, 
                 min_epsilon=.01, seed = 505):
        """
        Initialize variables, create grid for discretization.
        
        ARGUMENTS
        -----------------------
            - self: Reference to the current instance of the class
            - env: An instance of OpenAI Gym environment
            - state_grid: Starting grid which can be either continous or discretized
            - alpha: Learning rate
            - gamma: Discount factor
            - epsilon: Initial exploration rate
            - epsilon_decay_rate: Indicating how quickly decrease the epsilon
            - min_epsilon: minimum exploration rate
            - seed: random seed
        """
        # Environment info
        self.env = env
        self.state_grid = state_grid
        self.state_size = tuple(len(splits) + 1 for splits in self.state_grid) # n-dimensional state space
        self.action_size = self.env.action_space.n                             # 1-dimensional discrete action space
        self.seed = np.random.seed(seed)
        #print("Environment:", self.env)
        #print("State space size:", self.state_size)
        #print("Action space size:", self.action_size)
        
        # Learning parameters
        self.alpha = alpha  
        self.gamma = gamma  
        self.epsilon = self.initial_epsilon = epsilon 
        self.epsilon_decay_rate = epsilon_decay_rate 
        self.min_epsilon = min_epsilon
        
        # Create Q-table
        self.q_table = np.zeros(shape=(self.state_size + (self.action_size,)))
        #print("Q table size:", self.q_table.shape)

        
        
    def preprocess_state(self, state):
        """
        Map a continuous state to its discretized representation.
        
        ARGUMENTS
        ---------------------
            - self: Reference to the current instance of the class
            - state: State
            
        RETURNS
        ---------------------
            - discretized: Discretized representation
        """
        discretized = tuple(discretize(state, self.state_grid))
        return discretized

    
    
    def reset_episode(self, state):
        """
        Reset variables for a new episode.
        
        ARGUMENTS
        ---------------------
            - self: Reference to the current instance of the class
            - state: State
        
        RETURNS
        ---------------------
            - last_action: Last action
        """
        # Gradually decrease exploration rate
        self.epsilon *= self.epsilon_decay_rate
        self.epsilon = max(self.epsilon, self.min_epsilon)

        # Decide initial action
        self.last_state = self.preprocess_state(state)
        self.last_action = np.argmax(self.q_table[self.last_state])
        
        return self.last_action
    
    
    
    def reset_exploration(self, epsilon = None):
        """
        Reset exploration rate used when training.
        
        ARGUMENTS
        ---------------------
            - self: Reference to the current instance of the class
            - epsilon: Exploration rate
        """
        # Update the epsilon
        self.epsilon = epsilon if epsilon is not None else self.initial_epsilon

        
        
    def act(self, state, reward=None, done=None, mode='train'):
        """
        Pick next action and update internal Q table (when mode != 'test').
        
        ARGUMENTS
        ---------------------
            - self: Reference to the current instance of the class
            - state: State
            - reward: Reward
            - done: Indicates if its terminal state or not
            - mode: Can be use for training or testing
        
        RETURNS
        ---------------------
            - action: Action to take
        """
        # Map a continuous state to its discretized representation
        state = self.preprocess_state(state)
        
        # If testing
        if mode == 'test':
            
            # Produce an action
            action = np.argmax(self.q_table[state])
            
        # If training (this is default)
        else:
            
            # We update the Q table entry for the *last* (state, action) pair with current state, reward
            self.q_table[self.last_state + (self.last_action,)] += self.alpha * \
                (reward + self.gamma * max(self.q_table[state]) - self.q_table[self.last_state + (self.last_action,)])

            # Exploration vs. exploitation
            do_exploration = np.random.uniform(0, 1) < self.epsilon
            
            # If exploration
            if do_exploration:
                
                # Pick a random action
                action = np.random.randint(0, self.action_size)
                
            # If exploitation
            else:
                
                # Pick the best action from Q table
                action = np.argmax(self.q_table[state])

        # Roll over current state, action for next step
        self.last_state = state
        self.last_action = action
        
        return action
````

````python
# Initialize the Q-Learning agent (can act on a continuous state space by discretizing it.)
q_agent = QLearningAgent(env, state_grid)
````

Let's also define a convenience function to run an agent on a given environment.  When calling this function, you can pass in `mode='test'` to tell the agent not to learn.

````python
def run(agent, env, num_episodes=20000, mode='train'):
    """
    Run agent in given reinforcement learning environment and return scores.
    
    ARGUMENTS
    ---------------------
        - agent: The Q-Learning agent
        - env: An instance of OpenAI Gym environment
        - num_episodes: Number of episodes to run this game
        - mode: mode: Can be use for training or testing
    
    RETURNS
    ---------------------
        - scores: Cumulative reward
    """
    # Initialize an empty array for scores
    scores = []
    
    # Maximum average scores 
    max_avg_score = -np.inf
    
    # Iterate through episodes number
    for i_episode in range(1, num_episodes+1):
        
        # Resets the state and get an initial observation.
        state = env.reset()
        
        # Start action
        action = agent.reset_episode(state)
        
        # Initialize total reward
        total_reward = 0
        
        # Initialize the terminal state
        done = False

        # While not in terminal state
        while not done:
            
            # Run one timestep of the environment's dynamics
            state, reward, done, info = env.step(action)
            
            # Append the rewards to total rewards
            total_reward += reward
            
            # Choose an action
            action = agent.act(state, reward, done, mode)

        # Append total rewards into scores
        scores.append(total_reward)
        
        # If training
        if mode == 'train':
            
            # If less than 100 scores
            if len(scores) > 100:
                
                # Calculate the average scores
                avg_score = np.mean(scores[-100:])
                
                # If average scores bigger than maximum average scores
                if avg_score > max_avg_score:
                    
                    # Update the maximum average scores
                    max_avg_score = avg_score
                    
            # Every 100 episodes
            if i_episode % 100 == 0:
                
                # Print the stas
                print("\rEpisode {}/{} | Max Average Score: {}".format(i_episode, num_episodes, max_avg_score), end="")
                
                # Clean the prints
                sys.stdout.flush()

    return scores
````

````python
# Run agent in the environment and return its scores
scores = run(q_agent, env)
````

Now let's plot the scores. <i><strong>A good result is when scores increases as the agent goes through more episodes.</strong><i>

````python
# Visualize; The scores are noisy. That's why it's difficult to tell whether your agent is actually learning or not.
plt.plot(scores) 
plt.title("Scores per Episode")
````

<i><strong>In case our scores our noisy, we will use rolling mean to see if the agent is learning or not</strong><i>

````python
def plot_scores(scores, rolling_window=100):
    """
    Plot scores and optional rolling mean using specified window. 
    
    ARGUMENTS
    ------------------
        - scores
        - rolling_window
        
    RETURNS
    ------------------
        - rolling_mean
    """
    # Plot the scores
    plt.plot(scores) 
    
    # Title
    plt.title("Scores");
    
    # Calculate the rolloing mean
    rolling_mean = pd.Series(scores).rolling(rolling_window).mean()
    
    # Plot the rolling mean
    plt.plot(rolling_mean)
    
    return rolling_mean
````

````python
# Plot scores and optional rolling mean
rolling_mean = plot_scores(scores)
````

<i><strong>As you can see, the mean episode scores go up over time which is a good sign.</strong></i> Next, you can freeze learning and run the agent in test mode to see how well it performs.

````python
# Run the agent in test mode
test_scores = run(q_agent, env, num_episodes=100, mode='test')

# Print the scores and episodes
print("[TEST] Completed {} episodes with avg. score = {}".format(len(test_scores), np.mean(test_scores)))

# Visualize
_ = plot_scores(test_scores)
````

It's also interesting to look at the final Q-table that is learned by the agent. Note that the Q-table is of size MxNxA, where (M, N) is the size of the state space, and A is the size of the action space. We are interested in the maximum Q-value for each state, and the corresponding (best) action associated with that value.

````python
def plot_q_table(q_table):
    """
    Visualize max Q-value for each state and corresponding action.
    
    ARGUMENTS
    ------------------
        - q_table: Q-table with size of MxNxA, which (M, N) is the size of the state space, and A is 
                   the size of the action space.

    """
    # Max Q-value for each state
    q_image = np.max(q_table, axis=2)
    
    # Best action for each state
    q_actions = np.argmax(q_table, axis=2)  

    # Create a figure and a set of subplots
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Visualize max Q-value for each state
    cax = ax.imshow(q_image, cmap='jet');
    
    # Colorbar
    cbar = fig.colorbar(cax)
    
    # Iterate through first dimension of q_image
    for x in range(q_image.shape[0]):
        
        # Iterate through second dimension of q_image
        for y in range(q_image.shape[1]):
            
            # Set the text inside the plot
            ax.text(x, y, q_actions[x, y], color='white', horizontalalignment='center', verticalalignment='center')
            
    # Hide axes and gridlines
    ax.grid(False)
    
    # Title
    ax.set_title("Q-table, size: {}".format(q_table.shape))
    
    # X label
    ax.set_xlabel('position')
    
    # Y label
    ax.set_ylabel('velocity')
````

````python
# Visualize max Q-value for each state and corresponding action
plot_q_table(q_agent.q_table)
````
